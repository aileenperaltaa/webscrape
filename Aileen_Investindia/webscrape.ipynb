```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
source_code = requests.get('https://www.investindia.gov.in/bip/resources/list-ngos-providing-relief-during-covid-19').text
```

```python
soup = BeautifulSoup(source_code)
```

```python
arr = []
table_body=soup.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    arr.append(cols)
```

```python
arr
```

```python
column_names = arr[0]
ngo_name = []
ngo_description = []
ngo_state = []
ngo_contact = []
for x in arr[1:]:
    ngo_name.append(x[0])
    ngo_description.append(x[1])
    ngo_state.append(x[2])
    ngo_contact.append(x[3])
```

```python
import pandas as pd
data = {column_names[0]: ngo_name, column_names[1]: ngo_description, column_names[2]: ngo_state, column_names[3]: ngo_contact}
df = pd.DataFrame(data)
```

```python
df.to_csv('aileen_webscrape.csv')
```
